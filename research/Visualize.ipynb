```python
import pandas as pd
import matplotlib.pyplot as plt

# ==== Load your CSVs ====
df1 = pd.read_csv('weather_data.csv')
df2 = pd.read_csv('weather_data2023.csv')
# df3 = pd.read_csv('weather_data2024.csv')
df3 = pd.read_csv('Green Chilly Prices 21-25.csv')

# ==== Convert 'date' column to datetime ====
df1['date'] = pd.to_datetime(df1['date'])
df2['date'] = pd.to_datetime(df2['date'])
df3['date'] = pd.to_datetime(df3['date'])

# # ==== Normalize price column name (optional) ====
# # Rename if your price columns are named differently
# df1.rename(columns={'today': 'price', 'predicted_price': 'price'}, inplace=True)
# df2.rename(columns={'today': 'price', 'predicted_price': 'price'}, inplace=True)
# df3.rename(columns={'today': 'price', 'predicted_price': 'price'}, inplace=True)

# ==== Plot ====
plt.figure(figsize=(12, 6))

plt.plot(df1['date'], df1['today'], label='Dataset 1', marker='o')
# plt.plot(df1['date'], df1['dew'], label='Temp', marker='o')
plt.plot(df2['date'], df2['today'], label='Dataset 2', marker='s')
plt.plot(df3['date'], df3['today'], label='Dataset 3', marker='^')

plt.title('Price Comparison Over Time')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
